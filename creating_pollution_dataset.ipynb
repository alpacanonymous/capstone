```python
import pandas as pd
import numpy as np
```

I drop several columns for the following reasons.
>
> `Unnamed: 0` | No information explaining this column.
>
> `County Code` | Different counties in different states can have the same value, so I encode `County` later.
>
> `Site Num`, `Address`, & `State` | There are already multiple features regarding location. I am going to focus on `State Code`, `County`, and `City`.
>
> `NO2 Units`, `O3 Units`, `SO2 Units`, & `CO Units` | Units are not essential for my purposes.
>
> `NO2 1st Max Hour`, `O3 1st Max Hour`, `SO2 1st Max Hour`, & `CO 1st Max Hour` | Time of recording is not essential for my purposes.

## Ozone

```python
O3original = pd.read_csv('O3_2000_2021.csv')
```

```python
O3 = O3original.copy()

O3.drop(O3[(O3['State Name'] == 'Country Of Mexico') | 
           (O3['State Name'] == 'Virgin Islands') | 
           (O3['State Name'] == 'Canada')].index, inplace=True)

O3.drop(['State Code', 'County Code', 'Site Num', 'Parameter Code', 'POC', 'Latitude', 'Longitude', 'Datum',
         'Parameter Name', 'Sample Duration', 'Pollutant Standard', 'Units of Measure', 'Event Type', 
         'Observation Count', 'Observation Percent', 'Method Code', 'Method Name', 'Local Site Name',
         'CBSA Name', 'Date of Last Change'], axis=1, inplace=True)

O3.rename(columns={'Date Local': 'Date', 
                   'Arithmetic Mean': 'O3 Mean', 
                   '1st Max Value': 'O3 1st Max Value', 
                   '1st Max Hour': 'O3 1st Max Hour', 
                   'AQI': 'O3 AQI', 
                   'State Name': 'State', 
                   'County Name': 'County', 
                   'City Name': 'City'}, inplace=True)

O3['Date'] = pd.to_datetime(O3['Date'])

O3.info()
```

## Carbon Monoxide

```python
COoriginal = pd.read_csv('CO_2000_2021.csv')
```

```python
CO = COoriginal.copy()

CO.drop(CO[(CO['State Name'] == 'Country Of Mexico') | 
           (CO['State Name'] == 'Virgin Islands') | 
           (CO['State Name'] == 'Canada')].index, inplace=True)

CO.drop(CO[CO['Pollutant Standard'] == ('CO 1-hour 1971')].index, axis=0, inplace=True)

CO.drop(['State Code', 'County Code', 'Site Num', 'Parameter Code', 'POC', 'Latitude', 'Longitude', 'Datum',
         'Parameter Name', 'Sample Duration', 'Pollutant Standard', 'Units of Measure', 'Event Type', 
         'Observation Count', 'Observation Percent', 'Method Code', 'Method Name', 'Local Site Name',
         'CBSA Name', 'Date of Last Change'], axis=1, inplace=True)

CO.rename(columns={'Date Local': 'Date', 
                   'Arithmetic Mean': 'CO Mean', 
                   '1st Max Value': 'CO 1st Max Value', 
                   '1st Max Hour': 'CO 1st Max Hour', 
                   'AQI': 'CO AQI', 
                   'State Name': 'State', 
                   'County Name': 'County', 
                   'City Name': 'City'}, inplace=True)

CO['Date'] = pd.to_datetime(CO['Date'])

CO.info()
```

## Sulfur Dioxide

```python
SO2original = pd.read_csv('SO2_2000_2021.csv')
```

```python
SO2 = SO2original.copy()

SO2.drop(SO2[(SO2['State Name'] == 'Country Of Mexico') | 
             (SO2['State Name'] == 'Virgin Islands') | 
             (SO2['State Name'] == 'Canada')].index, inplace=True)

SO2.drop(SO2[SO2['Pollutant Standard'] == ('SO2 3-hour 1971')].index, inplace=True)

SO2.drop(['State Code', 'County Code', 'Site Num', 'Parameter Code', 'POC', 'Latitude', 'Longitude', 'Datum',
          'Parameter Name', 'Sample Duration', 'Pollutant Standard', 'Units of Measure', 'Event Type', 
          'Observation Count', 'Observation Percent', 'Method Code', 'Method Name', 'Local Site Name',
          'CBSA Name', 'Date of Last Change'], axis=1, inplace=True)

SO2.rename(columns={'Date Local': 'Date', 
                    'Arithmetic Mean': 'SO2 Mean', 
                    '1st Max Value': 'SO2 1st Max Value', 
                    '1st Max Hour': 'SO2 1st Max Hour', 
                    'AQI': 'SO2 AQI', 
                    'State Name': 'State', 
                    'County Name': 'County', 
                    'City Name': 'City'}, inplace=True)

SO2['Date'] = pd.to_datetime(SO2['Date'])

SO2.info()
```

## Nitrogen Dioxide

```python
NO2original = pd.read_csv('NO2_2000_2021.csv')
```

```python
NO2 = NO2original.copy()

NO2.drop(NO2[(NO2['State Name'] == 'Country Of Mexico') | 
             (NO2['State Name'] == 'Virgin Islands') | 
             (NO2['State Name'] == 'Canada')].index, inplace=True)

NO2.drop(['State Code', 'County Code', 'Site Num', 'Parameter Code', 'POC', 'Latitude', 'Longitude', 'Datum', 
          'Parameter Name', 'Sample Duration', 'Pollutant Standard', 'Units of Measure', 'Event Type', 
          'Observation Count', 'Observation Percent', 'Method Code', 'Method Name', 'Local Site Name', 
          'CBSA Name', 'Date of Last Change'], axis=1, inplace=True)

NO2.rename(columns={'Date Local': 'Date', 
                    'Arithmetic Mean': 'NO2 Mean', 
                    '1st Max Value': 'NO2 1st Max Value', 
                    '1st Max Hour': 'NO2 1st Max Hour', 
                    'AQI': 'NO2 AQI', 
                    'State Name': 'State', 
                    'County Name': 'County', 
                    'City Name': 'City'}, inplace=True)

NO2['Date'] = pd.to_datetime(NO2['Date'])

NO2.info()
```

## Pollution All

```python
O3df = O3.copy()
COdf = CO.copy()
SO2df = SO2.copy()
NO2df = NO2.copy()
```

```python
df1 = pd.merge(O3df, COdf, how='inner', 
               left_on=['Date', 'Address', 'State', 'County', 'City'], 
               right_on=['Date', 'Address', 'State', 'County', 'City'])
```

```python
df2 = pd.merge(df1, SO2df, how='inner', 
               left_on=['Date', 'Address', 'State', 'County', 'City'], 
               right_on=['Date', 'Address', 'State', 'County', 'City'])
```

```python
df3 = pd.merge(df2, NO2df, how='inner', 
               left_on=['Date', 'Address', 'State', 'County', 'City'], 
               right_on=['Date', 'Address', 'State', 'County', 'City'])
```

```python
df3.info()
```

```python
df_final = df3.copy()
df_final['Date'] = pd.to_datetime(df_final['Date'])
df_final['Day'] = df_final['Date'].dt.day
df_final['Month'] = df_final['Date'].dt.month
df_final['Year'] = df_final['Date'].dt.year
df_final
```

```python
df_final.columns
```

```python
reordered = ['Date', 'Year', 'Month', 'Day', 'Address', 'State', 'County', 'City', 
             'O3 Mean', 'O3 1st Max Value', 'O3 1st Max Hour', 'O3 AQI', 
             'CO Mean', 'CO 1st Max Value', 'CO 1st Max Hour', 'CO AQI', 
             'SO2 Mean', 'SO2 1st Max Value', 'SO2 1st Max Hour', 'SO2 AQI', 
             'NO2 Mean', 'NO2 1st Max Value', 'NO2 1st Max Hour', 'NO2 AQI']
```

```python
df_final = df_final.reindex(columns=reordered)
```
